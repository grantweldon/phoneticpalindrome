```python
def phonetic_palindrome(seed_word):
    # make input lowercase
    seed_word = seed_word.lower()
    
    # initialize phrase array
    phrases_temp = []
    
    # get phonemes of seed word
    phones = pronouncing.phones_for_word(seed_word)
    
    # loop over possible pronunciations of seed word
    for i in range(len(phones)):
        
        # here, we add any personal pronunciation changes we wish to make
        # our choice here gives a hard-e sound to words with "ea" that have an "ih" sound
        # for example, we give "dear" the pronunciation "deer" 
        if "IH1" in phones[i] and "ea" in seed_word:
            phones.append(  phones[i].replace("IH1", "IY1")   )
        
        # search phonemes of seed word for new first words containing seed phonemes (e.g. "leaky" from "leak")
        phones_cut = '^' + phones[i] # don't search for words that start with a different phoneme (e.g. we don't want "sleek" from "leak")
        first_words = pronouncing.search(phones_cut)
        
        # search for second words based on the first phoneme of the seed word (e.g. "tuba" -> boot)
        phones_spl = str.split(phones[i])
        first_phoneme_last = phones_spl[0] + '$'
        second_words = pronouncing.search(first_phoneme_last)
    
        # loop over possible first words
        for j in range(len(first_words)):
            phones_first = pronouncing.phones_for_word(first_words[j])
            
            # loop over possible pronunications of first words
            for k in range(len(phones_first)):
                phones_first_spl = str.split(phones_first[k])
                
                # loop over possible second words 
                for l in range(len(second_words)):
                    phones_second = pronouncing.phones_for_word(second_words[l])
                    
                    # loop over possible pronuncations of second words
                    for m in range(len(phones_second)):
                        phones_second_spl = str.split(phones_second[m])
                        
                        # combine phonemes of first and second words
                        phrase_spl = phones_first_spl + phones_second_spl
                        phrase_rev = phrase_spl[::-1]
                        
                        # save the phrase if same phonemes forward and backward
                        if ( phrase_spl == phrase_rev ):  
                            phrase = first_words[j] + ' ' + second_words[l]
                            phrases_temp.append(phrase)
    
    # delete duplicate phrases
    phrases = []
    [phrases.append(x) for x in phrases_temp if x not in phrases]
    
    # eliminate homophones (we don't want "adieu" if input "ado")
    phrases = [word for word in phrases if word.startswith(seed_word)]
    
    # message returned if no palindromes found
    if not phrases:
        phrases.append("I couldn't find any phonetic palindromes for that word :(")
    return phrases
```
